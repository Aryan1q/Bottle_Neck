```python
"""
╔══════════════════════════════════════════════════════════════════════╗
║  CONVOKE 8.0 — KnowledgeQuarry | PROBLEM_01: The Bottleneck Problem  ║
║  ML Engineering Track                                                ║                                                                     ║
╚══════════════════════════════════════════════════════════════════════╝

DATASET COLUMNS:
  vehicle_id       – unique vehicle identifier
  arrival_time_sec – when the vehicle arrives at the bottleneck (seconds)
  lane             – approach lane (1, 2, or 3)
  speed_kmph       – approach speed in km/h
  aggressiveness   – driver behaviour class: low | medium | high
  merge_delay_sec  – delay incurred during the merge/bottleneck event (seconds)
  exit_time_sec    – when the vehicle clears the bottleneck (seconds)

DERIVED TARGETS (computed in preprocessing):
  waiting_time     – exit_time_sec - arrival_time_sec  (total time in system)
  congestion_flag  – 1 if merge_delay_sec > median threshold (binary label)

ARCHITECTURE OVERVIEW (mirrors the original NST pipeline):
  ┌──────────────────────────────────────────────────────┐
  │  Data Ingest & Preprocessing  (≈ vgg_preprocess)     │
  │        ↓                                             │
  │  Feature Encoder (Dense encoder ≈ VGG encoder)       │
  │        ↓                                             │
  │  AdaIN-inspired Feature Fusion → Traffic Embedding   │
  │        ↓                                             │
  │  Decoder / Regressor  (merge_delay & waiting_time)   │
  │        ↓                                             │
  │  Loss Net (multi-task: MSE + classification)         │
  │        ↓                                             │
  │  Congestion Classifier                               │
  │        ↓                                             │
  │  Signal Optimiser  (rule-based + learned strategy)   │
  └──────────────────────────────────────────────────────┘
"""

# ─────────────────────────── Imports ────────────────────────────────
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, metrics, callbacks
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (mean_absolute_error, mean_squared_error,
                             classification_report, confusion_matrix)
from google.colab import files, drive

# ─────────────────────── Mount Drive ─────────────────────────────────
drive.mount('/content/drive')

# ─────────────────────── Global Config ───────────────────────────────
CHECKPOINT_PATH = "/content/drive/MyDrive/traffic_bottleneck/model_weights.weights.h5"
os.makedirs(os.path.dirname(CHECKPOINT_PATH), exist_ok=True)

BATCH_SIZE      = 32
EPOCHS          = 50
LATENT_DIM      = 64     # encoder output dimension (≈ 512 feature maps in VGG)
STYLE_WEIGHT    = 0.4    # weight for classification loss in multi-task loss
LEARNING_RATE   = 1e-3
RANDOM_STATE    = 42
CONGESTION_PCTILE = 60   # vehicles above this percentile of merge_delay are "congested"

tf.random.set_seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)

# ─────────────────────── 1. Data Loading ─────────────────────────────
print("Upload traffic_bottleneck_dataset.csv")
uploaded = files.upload()
csv_name = list(uploaded.keys())[0]

raw_df = pd.read_csv(csv_name)
print(f"\n✅ Loaded {len(raw_df)} rows | Columns: {list(raw_df.columns)}")
print(raw_df.head())

# ─────────────────────── 2. Preprocessing ────────────────────────────
# (Analogous to vgg_preprocess — normalise inputs to a stable range)

def preprocess(df: pd.DataFrame):
    df = df.copy()

    # --- Sort for temporal features ---
    df = df.sort_values(["lane", "arrival_time_sec"])

    # --- Derived features ---
    df["waiting_time"]  = df["exit_time_sec"] - df["arrival_time_sec"]
    df["net_delay"]     = df["merge_delay_sec"]
    df["speed_norm"]    = df["speed_kmph"] / df["speed_kmph"].max()

    # OLD (global gap) ❌
    # df["arrival_gap"] = df["arrival_time_sec"].diff().fillna(0)

    # ✅ NEW: lane-wise inter-vehicle gap
    df["lane_gap"] = df.groupby("lane")["arrival_time_sec"].diff().fillna(0)

    # ✅ NEW: rolling congestion window (last 5 vehicles in same lane)
    df["rolling_congestion"] = (
        df.groupby("lane")["merge_delay_sec"]
        .transform(lambda x: x.rolling(window=5, min_periods=1).mean())
    )

    # --- Binary congestion label ---
    threshold = np.percentile(df["merge_delay_sec"], CONGESTION_PCTILE)
    df["congestion_flag"] = (df["merge_delay_sec"] >= threshold).astype(int)

    # --- Encode categoricals ---
    agg_map = {"low": 0, "medium": 1, "high": 2}
    df["aggressiveness_enc"] = df["aggressiveness"].map(agg_map)

    lane_dummies = pd.get_dummies(df["lane"], prefix="lane")
    df = pd.concat([df, lane_dummies], axis=1)

    return df, threshold

df, congestion_threshold = preprocess(raw_df)
print(f"\n[CONFIG] Congestion threshold (merge_delay_sec ≥ {congestion_threshold:.1f}s)")
print(f"[CONFIG] Congested vehicles: {df['congestion_flag'].sum()} / {len(df)}")

# --- Feature / Target split ---
FEATURE_COLS = [
    "arrival_time_sec", "speed_norm", "aggressiveness_enc",
    "arrival_gap",
    "lane_1", "lane_2", "lane_3"
]
# Ensure all one-hot columns exist (in case some lanes missing)
for col in ["lane_1", "lane_2", "lane_3"]:
    if col not in df.columns:
        df[col] = 0

X = df[FEATURE_COLS].values.astype(np.float32)
y_delay   = df["merge_delay_sec"].values.astype(np.float32)   # regression target
y_waiting = df["waiting_time"].values.astype(np.float32)       # secondary regression
y_class   = df["congestion_flag"].values.astype(np.float32)    # classification target

# Scale features (analogous to normalising pixel values to [0,1])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X).astype(np.float32)

# Scale regression targets
delay_scaler   = StandardScaler()
waiting_scaler = StandardScaler()
y_delay_s   = delay_scaler.fit_transform(y_delay.reshape(-1,1)).ravel().astype(np.float32)
y_waiting_s = waiting_scaler.fit_transform(y_waiting.reshape(-1,1)).ravel().astype(np.float32)

# Train / validation split
(X_tr, X_val,
 yd_tr, yd_val,
 yw_tr, yw_val,
 yc_tr, yc_val) = train_test_split(
    X_scaled, y_delay_s, y_waiting_s, y_class,
    test_size=0.2, random_state=RANDOM_STATE, stratify=y_class
)

print(f"\n[DATA] Train: {X_tr.shape[0]} | Val: {X_val.shape[0]}")

# ─────────────────────── 3. Model Components ─────────────────────────
# Mirrors: get_encoder / get_decoder / get_loss_net from NST

def build_encoder(input_dim: int, latent_dim: int = LATENT_DIM):
    """
    Dense feature encoder — analogous to the VGG encoder in NST.
    Learns a rich latent representation of traffic state.
    """
    inp = layers.Input(shape=(input_dim,), name="traffic_input")
    x = layers.Dense(128, activation="relu", name="enc_fc1")(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(latent_dim, activation="relu", name="enc_fc2")(x)
    x = layers.BatchNormalization()(x)
    return keras.Model(inp, x, name="encoder")


def build_decoder(latent_dim: int = LATENT_DIM):
    """
    Decodes latent traffic embedding → predicted merge_delay (regression).
    Analogous to the convolutional decoder in NST.
    """
    inp = layers.Input(shape=(latent_dim,), name="latent_input")
    x = layers.Dense(64, activation="relu", name="dec_fc1")(inp)
    x = layers.Dropout(0.15)(x)
    x = layers.Dense(32, activation="relu", name="dec_fc2")(x)
    out = layers.Dense(1, activation="linear", name="merge_delay_pred")(x)
    return keras.Model(inp, out, name="decoder")


def build_congestion_classifier(latent_dim: int = LATENT_DIM):
    """
    Binary classifier on the latent embedding → congestion flag.
    Analogous to the 'loss_net' multi-layer feature extractor in NST,
    but here it produces a classification signal used in multi-task loss.
    """
    inp = layers.Input(shape=(latent_dim,), name="cls_input")
    x = layers.Dense(32, activation="relu")(inp)
    x = layers.Dropout(0.2)(x)
    out = layers.Dense(1, activation="sigmoid", name="congestion_pred")(x)
    return keras.Model(inp, out, name="congestion_classifier")


def build_waiting_decoder(latent_dim: int = LATENT_DIM):
    """
    Secondary decoder → predicted waiting_time.
    Analogous to AdaIN-adjusted secondary synthesis in NST.
    """
    inp = layers.Input(shape=(latent_dim,), name="wait_latent")
    x = layers.Dense(32, activation="relu")(inp)
    out = layers.Dense(1, activation="linear", name="waiting_pred")(x)
    return keras.Model(inp, out, name="waiting_decoder")


# ─────────────────────── 4. Main Model ───────────────────────────────
# Analogous to NeuralStyleTransfer(tf.keras.Model)

class TrafficBottleneckModel(tf.keras.Model):
    """
    Multi-task traffic bottleneck model.

    Forward pass:
        vehicle features
            → Encoder           (latent embedding)
            → Decoder           (merge_delay regression)
            → WaitingDecoder    (waiting_time regression)
            → CongestionClf     (binary congestion classification)

    Loss (analogous to content_loss + style_weight * style_loss):
        total_loss = delay_mse + waiting_mse + STYLE_WEIGHT * bce_congestion
    """
    def __init__(self, encoder, decoder, waiting_decoder,
                 classifier, style_weight=STYLE_WEIGHT, **kwargs):
        super().__init__(**kwargs)
        self.encoder         = encoder
        self.decoder         = decoder
        self.waiting_decoder = waiting_decoder
        self.classifier      = classifier
        self.style_weight    = style_weight

    def compile(self, optimizer, reg_loss_fn, cls_loss_fn):
        super().compile()
        self.optimizer    = optimizer
        self.reg_loss_fn  = reg_loss_fn   # MSE for regression (≈ content loss)
        self.cls_loss_fn  = cls_loss_fn   # BCE for classification (≈ style loss)

        self.total_loss_tracker   = metrics.Mean(name="loss")
        self.delay_loss_tracker   = metrics.Mean(name="delay_mse")
        self.waiting_loss_tracker = metrics.Mean(name="waiting_mse")
        self.cls_loss_tracker     = metrics.Mean(name="congestion_bce")
        self.cls_acc_tracker      = metrics.BinaryAccuracy(name="congestion_acc")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.delay_loss_tracker,
            self.waiting_loss_tracker,
            self.cls_loss_tracker,
            self.cls_acc_tracker,
        ]

    def call(self, inputs, training=False):
        z        = self.encoder(inputs, training=training)
        delay_p  = self.decoder(z, training=training)
        wait_p   = self.waiting_decoder(z, training=training)
        cong_p   = self.classifier(z, training=training)
        return delay_p, wait_p, cong_p

    def train_step(self, data):
        x, (y_delay, y_waiting, y_class) = data
        with tf.GradientTape() as tape:
            delay_p, wait_p, cong_p = self(x, training=True)

            # Content-equivalent: regression losses
            loss_delay   = self.reg_loss_fn(y_delay,   delay_p)
            loss_waiting = self.reg_loss_fn(y_waiting, wait_p)

            # Style-equivalent: classification loss
            loss_cls = self.cls_loss_fn(
                tf.expand_dims(y_class, -1), cong_p
            )

            total = loss_delay + loss_waiting + self.style_weight * loss_cls

        grads = tape.gradient(total, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))

        self.total_loss_tracker.update_state(total)
        self.delay_loss_tracker.update_state(loss_delay)
        self.waiting_loss_tracker.update_state(loss_waiting)
        self.cls_loss_tracker.update_state(loss_cls)
        self.cls_acc_tracker.update_state(
            tf.expand_dims(y_class, -1), cong_p
        )

        return {m.name: m.result() for m in self.metrics}

    def test_step(self, data):
        x, (y_delay, y_waiting, y_class) = data
        delay_p, wait_p, cong_p = self(x, training=False)

        loss_delay   = self.reg_loss_fn(y_delay,   delay_p)
        loss_waiting = self.reg_loss_fn(y_waiting, wait_p)
        loss_cls = self.cls_loss_fn(
            tf.expand_dims(y_class, -1), cong_p
        )
        total = loss_delay + loss_waiting + self.style_weight * loss_cls

        self.total_loss_tracker.update_state(total)
        self.delay_loss_tracker.update_state(loss_delay)
        self.waiting_loss_tracker.update_state(loss_waiting)
        self.cls_loss_tracker.update_state(loss_cls)
        self.cls_acc_tracker.update_state(
            tf.expand_dims(y_class, -1), cong_p
        )
        return {m.name: m.result() for m in self.metrics}


# ─────────────────────── 5. Training Monitor ─────────────────────────
# Analogous to TrainMonitor callback in NST

class TrafficMonitor(tf.keras.callbacks.Callback):
    """Plots per-epoch merge_delay predictions vs actuals on a small sample."""

    def __init__(self, X_sample, y_delay_sample, delay_scaler):
        super().__init__()
        self.X_sample      = X_sample
        self.y_true        = delay_scaler.inverse_transform(
            y_delay_sample.reshape(-1, 1)).ravel()
        self.delay_scaler  = delay_scaler
        self.epoch_losses  = []

    def on_epoch_end(self, epoch, logs=None):
        self.epoch_losses.append(logs.get("loss", 0))

        if (epoch + 1) % 10 == 0:
            preds_s, _, _ = self.model(self.X_sample, training=False)
            preds = self.delay_scaler.inverse_transform(
                preds_s.numpy()).ravel()

            fig, axes = plt.subplots(1, 2, figsize=(14, 4))

            axes[0].scatter(self.y_true, preds, alpha=0.6, color="steelblue", edgecolors="k", lw=0.4)
            mn, mx = min(self.y_true.min(), preds.min()), max(self.y_true.max(), preds.max())
            axes[0].plot([mn, mx], [mn, mx], "r--", lw=1.5, label="Perfect prediction")
            axes[0].set_xlabel("Actual merge_delay (s)")
            axes[0].set_ylabel("Predicted merge_delay (s)")
            axes[0].set_title(f"Merge Delay — Epoch {epoch+1}")
            axes[0].legend()

            axes[1].plot(self.epoch_losses, color="orange")
            axes[1].set_xlabel("Epoch")
            axes[1].set_ylabel("Total Loss")
            axes[1].set_title("Training Loss Curve")

            plt.tight_layout()
            plt.show()


# ─────────────────────── 6. Assemble & Train ─────────────────────────

if __name__ == "__main__":

    N_FEATURES = X_tr.shape[1]

    encoder    = build_encoder(N_FEATURES, LATENT_DIM)
    decoder    = build_decoder(LATENT_DIM)
    wait_dec   = build_waiting_decoder(LATENT_DIM)
    classifier = build_congestion_classifier(LATENT_DIM)

    model = TrafficBottleneckModel(encoder, decoder, wait_dec, classifier,
                                   style_weight=STYLE_WEIGHT)

    model.compile(
        optimizer  = keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        reg_loss_fn= keras.losses.MeanSquaredError(),
        cls_loss_fn= keras.losses.BinaryCrossentropy()
    )

    # Build graph (MANDATORY before weight loading, mirrors NST dummy forward pass)
    dummy_in = tf.zeros((1, N_FEATURES), dtype=tf.float32)
    model(dummy_in)
    print("\n[MODEL] Graph built.")
    encoder.summary()
    decoder.summary()

    # ── Optional: load previous checkpoint ──────────────────────────
    # model.load_weights(CHECKPOINT_PATH)
    # print("✅ Loaded previous weights")

    # ── Prepare tf.data datasets ────────────────────────────────────
    def make_dataset(X, yd, yw, yc, shuffle=True):
        ds = tf.data.Dataset.from_tensor_slices(
            (X, (yd, yw, yc))
        )
        if shuffle:
            ds = ds.shuffle(buffer_size=512)
        return ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    train_ds = make_dataset(X_tr, yd_tr, yw_tr, yc_tr, shuffle=True)
    val_ds   = make_dataset(X_val, yd_val, yw_val, yc_val, shuffle=False)

    # Pick small sample for monitor (≈ test batch in TrainMonitor)
    sample_idx  = np.random.choice(len(X_val), size=min(50, len(X_val)), replace=False)
    X_sample    = X_val[sample_idx]
    yd_sample   = yd_val[sample_idx]

    checkpoint_cb = callbacks.ModelCheckpoint(
        filepath        = CHECKPOINT_PATH,
        save_weights_only=True,
        monitor         = "val_loss",
        mode            = "min",
        save_best_only  = True,
        verbose         = 1
    )

    early_stop_cb = callbacks.EarlyStopping(
        monitor   = "val_loss",
        patience  = 10,
        restore_best_weights=True,
        verbose   = 1
    )

    reduce_lr_cb = callbacks.ReduceLROnPlateau(
        monitor  = "val_loss",
        factor   = 0.5,
        patience = 5,
        min_lr   = 1e-6,
        verbose  = 1
    )

    monitor_cb = TrafficMonitor(X_sample, yd_sample, delay_scaler)

    print("\n[TRAIN] Starting training...")
    history = model.fit(
        train_ds,
        validation_data = val_ds,
        epochs          = EPOCHS,
        callbacks       = [checkpoint_cb, early_stop_cb, reduce_lr_cb, monitor_cb]
    )

    print("\n✅ Training complete. Best weights saved to Drive.")


# ─────────────────────── 7. Evaluation & Inference ───────────────────

    print("\n[EVAL] Running evaluation on validation set...")

    delay_preds_s, wait_preds_s, cong_preds = model(X_val, training=False)

    # Inverse-transform to original scale
    delay_preds = delay_scaler.inverse_transform(delay_preds_s.numpy()).ravel()
    wait_preds  = waiting_scaler.inverse_transform(wait_preds_s.numpy()).ravel()
    y_delay_orig   = delay_scaler.inverse_transform(yd_val.reshape(-1,1)).ravel()
    y_waiting_orig = waiting_scaler.inverse_transform(yw_val.reshape(-1,1)).ravel()

    cong_binary = (cong_preds.numpy().ravel() > 0.5).astype(int)

    # Regression metrics
    delay_mae  = mean_absolute_error(y_delay_orig,   delay_preds)
    delay_rmse = np.sqrt(mean_squared_error(y_delay_orig, delay_preds))
    wait_mae   = mean_absolute_error(y_waiting_orig, wait_preds)
    wait_rmse  = np.sqrt(mean_squared_error(y_waiting_orig, wait_preds))

    print(f"\n{'='*50}")
    print(f"  MERGE DELAY  — MAE: {delay_mae:.2f}s  |  RMSE: {delay_rmse:.2f}s")
    print(f"  WAITING TIME — MAE: {wait_mae:.2f}s   |  RMSE: {wait_rmse:.2f}s")
    print(f"\n  CONGESTION CLASSIFIER:")
    print(classification_report(yc_val.astype(int), cong_binary,
                                target_names=["No Congestion", "Congested"]))
    print(f"{'='*50}")


# ─────────────────────── 8. Visualisations ───────────────────────────

    fig = plt.figure(figsize=(18, 12))
    gs  = gridspec.GridSpec(2, 3, figure=fig, hspace=0.4, wspace=0.35)

    # (a) Actual vs Predicted merge_delay
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.scatter(y_delay_orig, delay_preds, alpha=0.55, color="steelblue", edgecolors="k", lw=0.3)
    mn, mx = min(y_delay_orig.min(), delay_preds.min()), max(y_delay_orig.max(), delay_preds.max())
    ax1.plot([mn, mx], [mn, mx], "r--", lw=1.5)
    ax1.set_xlabel("Actual Merge Delay (s)"); ax1.set_ylabel("Predicted (s)")
    ax1.set_title("Merge Delay — Actual vs Predicted")

    # (b) Actual vs Predicted waiting_time
    ax2 = fig.add_subplot(gs[0, 1])
    ax2.scatter(y_waiting_orig, wait_preds, alpha=0.55, color="darkorange", edgecolors="k", lw=0.3)
    mn2, mx2 = min(y_waiting_orig.min(), wait_preds.min()), max(y_waiting_orig.max(), wait_preds.max())
    ax2.plot([mn2, mx2], [mn2, mx2], "r--", lw=1.5)
    ax2.set_xlabel("Actual Waiting Time (s)"); ax2.set_ylabel("Predicted (s)")
    ax2.set_title("Waiting Time — Actual vs Predicted")

    # (c) Training loss curve
    ax3 = fig.add_subplot(gs[0, 2])
    ax3.plot(history.history["loss"], label="Train Loss", color="royalblue")
    ax3.plot(history.history["val_loss"], label="Val Loss", color="tomato")
    ax3.set_xlabel("Epoch"); ax3.set_ylabel("Loss")
    ax3.set_title("Multi-Task Training Loss"); ax3.legend()

    # (d) Merge delay distribution by aggressiveness
    df_val = df.iloc[X_val.shape[0]*-1:].copy() if len(df) > len(X_val) else df.copy()
    ax4 = fig.add_subplot(gs[1, 0])
    for agg, color in zip(["low","medium","high"], ["green","orange","red"]):
        subset = df[df["aggressiveness"] == agg]["merge_delay_sec"]
        ax4.hist(subset, bins=15, alpha=0.55, label=agg.capitalize(), color=color)
    ax4.axvline(congestion_threshold, color="black", linestyle="--", label=f"Threshold ({congestion_threshold:.1f}s)")
    ax4.set_xlabel("Merge Delay (s)"); ax4.set_ylabel("Vehicle Count")
    ax4.set_title("Merge Delay by Driver Aggressiveness"); ax4.legend()

    # (e) Throughput per lane (vehicles exiting per time window)
    ax5 = fig.add_subplot(gs[1, 1])
    df["time_bucket"] = pd.cut(df["exit_time_sec"], bins=10)
    lane_throughput = df.groupby(["time_bucket","lane"]).size().unstack(fill_value=0)
    lane_throughput.plot(ax=ax5, kind="bar", stacked=True,
                         color=["steelblue","darkorange","seagreen"], edgecolor="k", lw=0.3)
    ax5.set_xlabel("Exit Time Window"); ax5.set_ylabel("Vehicles Exited")
    ax5.set_title("Throughput per Lane over Time")
    ax5.tick_params(axis='x', rotation=45)
    ax5.legend(title="Lane")

    # (f) Congestion confusion matrix
    ax6 = fig.add_subplot(gs[1, 2])
    cm = confusion_matrix(yc_val.astype(int), cong_binary)
    im = ax6.imshow(cm, cmap="Blues")
    for i in range(2):
        for j in range(2):
            ax6.text(j, i, str(cm[i,j]), ha="center", va="center",
                     fontsize=14, color="black")
    ax6.set_xticks([0,1]); ax6.set_xticklabels(["No Cong.","Congested"])
    ax6.set_yticks([0,1]); ax6.set_yticklabels(["No Cong.","Congested"])
    ax6.set_xlabel("Predicted"); ax6.set_ylabel("Actual")
    ax6.set_title("Congestion Classification — Confusion Matrix")
    plt.colorbar(im, ax=ax6)

    plt.suptitle("CONVOKE 8.0 | PROBLEM_01: Traffic Bottleneck — Model Dashboard",
                 fontsize=14, fontweight="bold", y=1.01)
    plt.savefig("/content/traffic_dashboard.png", bbox_inches="tight", dpi=150)
    plt.show()
    print("✅ Dashboard saved: /content/traffic_dashboard.png")


# ─────────────────────── 9. Signal Optimiser ─────────────────────────
# Uses model predictions to recommend signal timing adjustments.
# (Analogous to the stylized-image output in NST — the "useful artifact"
#  produced by the model.)

    def optimise_signal_timing(model, scaler, delay_scaler, n_scenarios=6):
        """
        For different hypothetical traffic scenarios (arrival rate × aggressiveness),
        predict congestion and recommend green-phase duration adjustments.

        Strategy:
          - Predicted merge_delay > threshold → extend green phase for that lane
          - Low aggressiveness vehicles → can tolerate shorter phases
        """
        print("\n[SIGNAL OPTIMISER] Scenario Analysis")
        print("="*55)

        scenarios = []
        for lane in [1, 2, 3]:
            for agg_val, agg_label in enumerate(["low","medium","high"]):
                # Create a synthetic vehicle at a moderately busy arrival time
                feat = np.zeros((1, N_FEATURES), dtype=np.float32)
                # arrival_time_sec, speed_norm, aggressiveness_enc, arrival_gap, lane_1,2,3
                feat[0, 0] = 300.0          # arrival at 300s into simulation
                feat[0, 1] = 0.65           # moderate speed
                feat[0, 2] = float(agg_val) # aggressiveness
                feat[0, 3] = 15.0           # 15s gap since last vehicle
                feat[0, 4 + (lane-1)] = 1.0 # one-hot lane

                feat_s = scaler.transform(feat).astype(np.float32)
                d_pred_s, _, c_pred = model(feat_s, training=False)
                d_pred = delay_scaler.inverse_transform(d_pred_s.numpy())[0,0]
                c_prob = c_pred.numpy()[0,0]

                # Recommend: base green phase 30s; add 1s per 1s of predicted delay above 10s
                base_green   = 30
                extra_green  = max(0, int((d_pred - 10) * 0.8))
                recommended  = base_green + extra_green

                scenarios.append({
                    "Lane":         lane,
                    "Aggressiveness": agg_label,
                    "Pred Delay (s)": round(d_pred, 1),
                    "Cong. Prob":    round(float(c_prob), 3),
                    "Rec. Green (s)": recommended
                })

        results_df = pd.DataFrame(scenarios)
        print(results_df.to_string(index=False))

        # Bar chart of recommended green phase per scenario
        fig, ax = plt.subplots(figsize=(12, 5))
        labels = [f"L{r['Lane']}/{r['Aggressiveness']}" for _, r in results_df.iterrows()]
        colors = ["green" if p < 0.5 else "red" for p in results_df["Cong. Prob"]]
        bars = ax.bar(labels, results_df["Rec. Green (s)"], color=colors, alpha=0.75, edgecolor="k")
        ax.axhline(30, color="gray", linestyle="--", lw=1.2, label="Base Green (30s)")
        ax.set_xlabel("Scenario (Lane / Aggressiveness)")
        ax.set_ylabel("Recommended Green Phase (s)")
        ax.set_title("Signal Optimiser — Recommended Green Phase per Scenario\n"
                     "(Green = No Congestion Predicted | Red = Congestion Likely)")
        ax.legend()
        plt.tight_layout()
        plt.savefig("/content/signal_optimiser.png", bbox_inches="tight", dpi=150)
        plt.show()
        print("✅ Signal optimiser chart saved.")
        return results_df

    signal_df = optimise_signal_timing(model, scaler, delay_scaler)


# ─────────────────────── 10. Inference on New Data ───────────────────
# Analogous to the inference / testing section of the NST script

    def run_inference(model, scaler, delay_scaler, waiting_scaler, new_df=None):
        """
        Run predictions on a new CSV.  If none provided, uses last 10 rows of
        the training dataset as a demo.
        """
        if new_df is None:
            print("\n[INFERENCE] Using last 10 rows of original dataset as demo.")
            new_df = raw_df.tail(10).copy()

        new_proc, _ = preprocess(new_df)
        for col in ["lane_1","lane_2","lane_3"]:
            if col not in new_proc.columns:
                new_proc[col] = 0

        X_new = new_proc[FEATURE_COLS].values.astype(np.float32)
        X_new_s = scaler.transform(X_new).astype(np.float32)

        d_s, w_s, c_p = model(X_new_s, training=False)
        d_pred = delay_scaler.inverse_transform(d_s.numpy()).ravel()
        w_pred = waiting_scaler.inverse_transform(w_s.numpy()).ravel()
        cong   = (c_p.numpy().ravel() > 0.5).astype(int)

        out = new_df[["vehicle_id","lane","aggressiveness","merge_delay_sec"]].copy()
        out["pred_delay_s"]   = np.round(d_pred, 2)
        out["pred_waiting_s"] = np.round(w_pred, 2)
        out["congestion_pred"] = ["CONGESTED" if c else "CLEAR" for c in cong]

        print("\n[INFERENCE] Predictions:")
        print(out.to_string(index=False))
        return out

    inference_results = run_inference(model, scaler, delay_scaler, waiting_scaler)

    print("\n" + "="*60)
    print("  CONVOKE 8.0 — PROBLEM_01 COMPLETE")
    print("  Metrics: Avg Waiting Time ✓ | Throughput ✓ | Congestion ✓")
    print("="*60)
```
